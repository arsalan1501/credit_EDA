#### Converting days into absolute years for analysis (since values are incorrectly negative also)

```python
DaystoYrs = application_data.columns[application_data.columns.str.contains('DAYS_')].tolist()
DaystoYrs
```

```python
application_data[DaystoYrs] = round(application_data[DaystoYrs].abs()/365,0)
application_data.head()
```

#### univariate analysis of Days columns which we converted to yrs to look for outliers

```python
data = pd.DataFrame({"Age": application_data.DAYS_BIRTH, "Employment_yrs": application_data.DAYS_EMPLOYED,
                    "Registration": application_data.DAYS_REGISTRATION, "ID_Publish": application_data.DAYS_ID_PUBLISH})

ax = data[['Age', 'Employment_yrs', 'Registration', 'ID_Publish']].plot(kind='box', title='Outlier analysis on year columns')
```

```python
application_data.DAYS_EMPLOYED.describe()
```

Looks like the DAYS_EMPLOYED column has outliers/missing values since above 100yrs is incorrect. Lets see % of such values and treat them accordingly

```python
(application_data.DAYS_EMPLOYED[application_data.DAYS_EMPLOYED>100].count()/application_data.DAYS_EMPLOYED.count())*100
```

18% of this column has a value of more than 100yrs. Since the number of rows is high but not high enough to drop we can replace with blanks

```python
application_data.loc[application_data.DAYS_EMPLOYED>100,"DAYS_EMPLOYED"]=np.NaN
application_data.DAYS_EMPLOYED.describe()
```

Converting age in years to categorical

```python
application_data["DAYS_BIRTH_cat"]=pd.cut(application_data.DAYS_BIRTH,[0, 25, 35, 45, 55, 65, 999], labels= ["<25","25-35","35-45","45-55","55-65","65+"])
application_data["DAYS_BIRTH_cat"].value_counts(normalize= True)*100
```

```python
plt.figure(figsize=[10,3])
plt.subplot(1, 3, 1)
plt.title('Applications by age')
application_data.DAYS_BIRTH_cat.value_counts(normalize= True).plot.bar()
plt.subplot(1, 3, 2)
plt.title('Defaults by age')
application_data.groupby(['DAYS_BIRTH_cat'])['TARGET'].sum().plot.bar()
plt.subplot(1, 3, 3)
plt.title('%Defaults by age')
application_data.groupby(['DAYS_BIRTH_cat'])['TARGET'].mean().plot.bar();
```

Most applications are from 35-45 age group and most defaults are from 25-35 followed by 35-45. %defaults are most in the <25 age group

```python
application_data["DAYS_EMPLOYED_cat"]=pd.cut(application_data.DAYS_EMPLOYED,[0, 10, 20, 30, 50, 999], labels= ["<10","10-20","20-30","30-50", "50+"])
application_data["DAYS_EMPLOYED_cat"].value_counts(normalize= True)*100
```
